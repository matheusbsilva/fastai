```python
from duckduckgo_search import DDGS
import fastcore.all as core
from fastdownload import download_url
import fastai.vision.all as vision
```

# Data exploration

```python
def search_images(term: str, max_images: int=30):
    print(f"Searching for '{term}'")
    ddg = DDGS()
    return core.L(ddg.images(term, max_results=max_images)).itemgot('image')
```

```python
urls = search_images("cat")
```

```python
dest = 'data/cat.jpg'
download_url(urls[0], dest)
im = vision.Image.open(dest)
im.to_thumb(256, 256)
```

```python
download_url(search_images('tiger photos', max_images=1)[0], 'data/tiger.jpg')
vision.Image.open('data/tiger.jpg').to_thumb(256, 256)
```
